```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
x1=np.arange(-10,10,0.1)#arange-10100.1
x2=np.arange(-10,10,0.1)

X1,X2=np.meshgrid(x1,x2)
z=X1+X2-1
h= sigmoid(z)

fig=plt.figure(figsize=(5,4))
ax=plt.axes(projection='3d')
ax.plot_surface(X1,X2,h)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
